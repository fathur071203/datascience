### Feature Engginering

```python
# 1. Rasio Kelulusan Semester Pertama (Pass_Rate_1st_Sem)
data_df['Pass_Rate_1st_Sem'] = data_df['Curricular_units_1st_sem_approved'] / data_df['Curricular_units_1st_sem_enrolled']
data_df['Pass_Rate_1st_Sem'] = data_df['Pass_Rate_1st_Sem'].replace([np.inf, -np.inf], np.nan).fillna(0)

# 2. Indikator Risiko Finansial (Financial_Risk)
data_df['Financial_Risk'] = ((data_df['Debtor'].astype(int) == 1) | (data_df['Tuition_fees_up_to_date'].astype(int) == 0)).astype(int)

# 3. Tingkat Kualifikasi Pendidikan Orang Tua Tertinggi (Parents_Highest_Qualification)
# Konversi ke numerik, Unknown/Can't read or write (kode 34/35) dianggap terendah
data_df['Mothers_qualification_num'] = pd.to_numeric(data_df['Mothers_qualification'], errors='coerce').fillna(0)
data_df['Fathers_qualification_num'] = pd.to_numeric(data_df['Fathers_qualification'], errors='coerce').fillna(0)
data_df['Parents_Highest_Qualification'] = data_df[['Mothers_qualification_num', 'Fathers_qualification_num']].max(axis=1)

# Hapus kolom-kolom asli yang sudah direpresentasikan oleh fitur baru
data_df = data_df.drop([
    'Curricular_units_1st_sem_approved', 'Curricular_units_1st_sem_enrolled',
    'Debtor', 'Tuition_fees_up_to_date',
    'Mothers_qualification', 'Fathers_qualification',
    'Mothers_qualification_num', 'Fathers_qualification_num'
], axis=1)
```

**Insight:** rasio kelulusan semester pertama (Pass_Rate_1st_Sem) secara langsung menunjukkan seberapa baik kinerja akademik mahasiswa, sebuah indikator dropout yang jauh lebih jelas. Demikian pula, menggabungkan status utang dan keterlambatan pembayaran menjadi satu indikator risiko finansial (Financial_Risk) memberikan gambaran masalah keuangan yang lebih komprehensif. Terakhir, menyatukan kualifikasi pendidikan orang tua menjadi tingkat kualifikasi pendidikan orang tua tertinggi (Parents_Highest_Qualification) menyederhanakan informasi latar belakang keluarga yang bisa memengaruhi dukungan studi mahasiswa.

## Modeling

### Encoding

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()


valid_categorical_columns = [col for col in categorical_columns if col in data_df.columns]

for col in valid_categorical_columns:
    data_df[col] = label_encoder.fit_transform(data_df[col])

data_df
```

### Splitting

```python
X = data_df.drop('Status', axis=1)
y = data_df['Status']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

### Normalisasi

```python
numeric_columns = X.select_dtypes(include='number').columns

scaler = MinMaxScaler()
X_train[numeric_columns] = scaler.fit_transform(X_train[numeric_columns])
X_test[numeric_columns] = scaler.transform(X_test[numeric_columns])
```

```python
data_df
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier

# Inisialisasi model
models = {
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42),
    "Logistic Regression": LogisticRegression(random_state=42, max_iter=1000),
    "Random Forest": RandomForestClassifier(random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(random_state=42),
    "SVM": SVC(probability=True, random_state=42),
    "KNN": KNeighborsClassifier()
}

# Melatih dan mengevaluasi model
results = {}
for model_name, model in models.items():
    print(f"Training {model_name}...")
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_pred_proba = model.predict_proba(X_test)[:, 1] if hasattr(model, "predict_proba") else None

    # Evaluasi metrik
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='macro')
    recall = recall_score(y_test, y_pred, average='macro')
    f1 = f1_score(y_test, y_pred, average='macro')
    # For multiclass, roc_auc_score needs special handling
    if y_pred_proba is not None and len(np.unique(y_test)) > 2:
        # One-vs-rest multiclass ROC AUC
        roc_auc = roc_auc_score(y_test, model.predict_proba(X_test), multi_class='ovr', average='macro')
    elif y_pred_proba is not None:
        roc_auc = roc_auc_score(y_test, y_pred_proba)
    else:
        roc_auc = "N/A"

    # Menyimpan hasil
    results[model_name] = {
        "Accuracy": accuracy,
        "Precision": precision,
        "Recall": recall,
        "F1-Score": f1,
        "ROC AUC": roc_auc
    }

results_df = pd.DataFrame(results).T  # Transpose untuk menampilkan model sebagai baris
results_df.index.name = "Model"
results_df.reset_index(inplace=True)
results_df
```

## Evaluation

```python
from sklearn.preprocessing import label_binarize

# Function to plot ROC curve for multiclass
def plot_roc_curve_multiclass(y_true, y_proba, title='ROC Curve'):
    n_classes = y_proba.shape[1]
    y_true_bin = label_binarize(y_true, classes=np.arange(n_classes))
    plt.figure(figsize=(6, 4))
    for i in range(n_classes):
        fpr, tpr, _ = roc_curve(y_true_bin[:, i], y_proba[:, i])
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, lw=2, label=f'Class {i} (AUC = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], color='gray', lw=1, linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(title)
    plt.legend(loc="lower right")
    plt.grid(alpha=0.3)
    plt.show()

# Function to generate evaluation report
def evaluation_report(y_true, y_predicted, y_proba=None, title='Data'):
    print('Classification Report on ' + title)
    print(classification_report(y_true, y_predicted))

    plt.figure(figsize=(4, 3))
    sns.heatmap(confusion_matrix(y_true, y_predicted), annot=True, fmt='d', cmap='Blues')
    plt.title('Confusion Matrix on ' + title)
    plt.ylabel('True Label', fontsize=10)
    plt.xlabel('Predicted Label', fontsize=10)
    plt.yticks(fontsize=8)
    plt.xticks(fontsize=8)
    plt.show()

    # Plot ROC curve for multiclass if proba is available
    if y_proba is not None and len(np.unique(y_true)) > 2:
        plot_roc_curve_multiclass(y_true, y_proba, title=f'ROC Curve on {title}')
    elif y_proba is not None:
        # Binary case
        fpr, tpr, _ = roc_curve(y_true, y_proba)
        roc_auc = auc(fpr, tpr)
        plt.figure(figsize=(6, 4))
        plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
        plt.plot([0, 1], [0, 1], color='gray', lw=1, linestyle='--')
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title(f'ROC Curve on {title}')
        plt.legend(loc="lower right")
        plt.grid(alpha=0.3)
        plt.show()

# Evaluate all models
for model_name, model in models.items():
    print(f"Evaluation for {model_name} on Train Data")
    y_train_pred = model.predict(X_train)
    y_train_proba = model.predict_proba(X_train) if hasattr(model, "predict_proba") else None
    evaluation_report(y_train, y_train_pred, y_train_proba, title=f'{model_name} Train Data')

    print(f"Evaluation for {model_name} on Test Data")
    y_test_pred = model.predict(X_test)
    y_test_proba = model.predict_proba(X_test) if hasattr(model, "predict_proba") else None
    evaluation_report(y_test, y_test_pred, y_test_proba, title=f'{model_name} Test Data')
```

**Insight:** Berdasarkan hasil evaluasi model pada data uji, Gradient Boosting menonjol sebagai model dengan performa terbaik. Model ini menunjukkan akurasi tertinggi secara keseluruhan (0.78) dan nilai ROC AUC tertinggi (0.882), mengindikasikan kemampuan diskriminasi yang sangat baik di antara kelas-kelas. Lebih lanjut, untuk kelas 'Dropout' (prioritas utama Jaya Jaya Institut), Gradient Boosting mencapai precision tertinggi (0.83) dan recall yang kuat (0.77), menghasilkan F1-Score 0.80. Meskipun Logistic Regression juga memiliki F1-Score 0.80 untuk kelas 'Dropout' dengan recall sedikit lebih tinggi (0.80), precision Gradient Boosting yang lebih tinggi (0.83 vs 0.80) menunjukkan keseimbangan yang lebih baik antara identifikasi dropout yang tepat dan meminimalkan false positives. Dengan performa yang konsisten di berbagai metrik kunci dan kemampuannya menangani kompleksitas data, Gradient Boosting adalah pilihan yang paling efektif untuk memprediksi dropout mahasiswa.
